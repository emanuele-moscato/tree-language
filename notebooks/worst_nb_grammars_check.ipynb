```python
check_data = []

i = epsilons.shape[0] - 1

epsilon = epsilons[i]

for j, sigma in enumerate(sigmas):
    print(f'i = {i} | j = {j}')
    
    seed = seeds[i, j]

    bp_accuracy = bp_accuracies[epsilon_index, sigma_index]
    nb_accuracy = nb_accuracies[epsilon_index, sigma_index]

    rho, roots_train, leaves_train, roots_test, leaves_test = generate_dataset_jerome(
        seed=seed,
        q=q,
        k=k,
        epsilon=epsilon,
        sigma=sigma,
        n_samples_training=4000,
        n_samples_test=4000
    )

    test_data_in_training_frac = np.mean([
        (leaves_train == leaves_test[i, :]).all(axis=1).any()
        for i in range(leaves_test.shape[0])
    ])

    check_data.append({
        'q': q,
        'k': k,
        'epsilon': epsilon,
        'sigma': sigma,
        'bp_accuracy': bp_accuracy,
        'nb_accuracy': nb_accuracy,
        'test_data_in_training_frac': test_data_in_training_frac
    })

check_data = pd.DataFrame(check_data)
```
